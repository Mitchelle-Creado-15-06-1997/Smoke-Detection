# 2. Data preparation :

## 2.1 Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
```

## 2.2 Load data

```python
Data = pd.read_csv('./smoke.csv')
Data.head(5000)
```

## 2.3 Describtion

```python
Data['Fire Alarm'].describe()
```

## 2.4 Check for null

```python
total = Data.isnull().sum().sort_values(ascending=False)
percent=(Data.isnull().sum()/Data.isnull().count()).sort_values(ascending=False)
missingData = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
missingData.head(15)
```

## 2.5 Get target

```python
X = Data.iloc[:,:-1]
Y = Data.iloc[:,-1]
```

## KFold and Gradient boosting classifier


#### Gradient Boosting Classifier:

Gradient boosting classifier is a machine learning technique used in regression and classification tasks among others it gives a predict model in the form of an ensemble of weak prediction models .


--------------------------------------------------------------------------------------------------------------

#### KFold:

KFolds cross validation 
Provides train/test indices to split data train/test. Split dataset into k consecutive folds (without shuffling by default).
Each fold is then used once as a vlidation while the k-1 remining folds form the trainig set.

<center><img src ="https://upload.wikimedia.org/wikipedia/commons/b/b5/K-fold_cross_validation_EN.svg" /></center>

```python
KFModel = StratifiedKFold(n_splits=5)
p_rf= np.zeros(Y.shape[0])
```

```python
for n in range(10,100,10):
    for train,test in KFModel.split(X,Y):
        X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=33)
        clf = GradientBoostingClassifier(n_estimators=100,learning_rate=1.0,max_depth=1,random_state=33)
        clf.fit(X_train,Y_train)
        Y_pred = clf.predict(X_test)
        cm=confusion_matrix(Y_test,Y_pred)
        print(cm)
    print('Score for n:','',accuracy_score(Y_test,Y_pred))
```
